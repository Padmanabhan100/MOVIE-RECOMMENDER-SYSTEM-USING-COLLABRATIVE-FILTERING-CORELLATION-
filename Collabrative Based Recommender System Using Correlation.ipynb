###### RENAMING COLUMN TO ID TO MAPP THE VALUES OF 2 DAATFRAMES

```python
ratings.rename(columns={"movieId":"id"},inplace=True)
```

```python
# converting id to strings 
ratings['id'] = ratings['id'].astype("str")
movies['id'] = movies['id'].astype("str")
```

```python
# merging df to getvthe names of the movies
ratings_merged = pd.merge(ratings,movies[['title','id','vote_count']],on="id")
```

```python
# checking the dataframe
ratings_merged.sort_values("rating",ascending=False).head()
```

```python
user_ratings = ratings_merged.sort_values("rating",ascending=False)
```

```python
user_ratings.reset_index(inplace=True)
```

```python
del user_ratings['index']
```

```python
user_ratings.head()
```

#### NOW THAT OUR USER RATINGS DATA IS SET,LETS MAKE A PIVOT TABLE WHICH SHOWS RATINGS OF EACH USER

```python
user_mat = pd.pivot_table(data=user_ratings,index='userId',columns="title",values="rating")
user_mat.head()
```

#### MAKING A FUNCTION THAT RETURNS SUGGESTION

```python
def suggest(movie_name):
    # RATING OF GIVEN MOVIE FROM PIVOT TABLE
    user_movie_rating = user_mat[movie_name]
    # MAKING A DATARAME OF CORELLATION VALUES OF GIVEN MOVIE BY COMPARING WITH THE PIVOT
    similar_to_movie = pd.DataFrame(user_mat.corrwith(user_movie_rating),columns=['corellation'])
    # MERGING VOTE COUNTS OF MOVIES AND TITLE
    similar_movies = similar_to_movie.merge(movies[["vote_count",'title']],on='title')
    # RETURNING RESULTS
    return similar_movies[similar_movies['corellation']>=0.8].sort_values("vote_count",ascending=False)
```

```python
suggest("Hulk")
```

* THE RECOMMENDATIONS ARE NOT RELATED TO THE GENRES
